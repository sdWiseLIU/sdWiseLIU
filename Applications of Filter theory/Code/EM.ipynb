```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
f = pd.read_csv('/Users/apple/Desktop/important file/Stochastic Control/Code of Final project/EM/3.csv', header = None)
data = pd.DataFrame(f)
data = data.values
f_1 = pd.read_csv('/Users/apple/Desktop/important file/Stochastic Control/Code of Final project/EM/4.csv', header = None)
data_1 = pd.DataFrame(f_1)
data_1 = data_1.values
```

```python
a_low = data[:,0]
a_high = data[:,1]
b_low = data[:,2]
b_high = data[:,3]
c_low = data[:,4]
c_high = data[:,5]
a_av = data_1[:,0]
b_av = data_1[:,1]
c_av = data_1[:,2]
```

```python
time = 10*np.linspace(1, 10, 10)
```

```python
plt.plot(time, a_low, ls='--')
plt.plot(time, a_high, ls='-.')
plt.plot(time, a_av, label = 'avarge of a', ls='-', marker='o', color='purple')
plt.fill_between(time, a_low, a_high, color='pink', alpha=0.3)
plt.xlabel('Day')
plt.ylabel('Estimation value')
plt.title('95% confidence interval of a')
plt.legend()
plt.savefig("EM_a.png", dpi=500, bbox_inches = 'tight')
plt.show()
```

```python
plt.plot(time, b_low, ls='--')
plt.plot(time, b_high, ls='-.')
plt.plot(time, b_av, label = 'avarge of b', ls='-', marker='o', color='purple')
plt.fill_between(time, b_low, b_high, color='pink', alpha=0.3)
plt.xlabel('Day')
plt.ylabel('Estimation value')
plt.title('95% confidence interval of b')
plt.legend()
plt.savefig("EM_b.png", dpi=500, bbox_inches = 'tight')
plt.show()
```

```python
plt.plot(time, c_low, ls='--')
plt.plot(time, c_high, ls='-.')
plt.plot(time, c_av, label = 'avarge of c', ls='-', marker='o', color='purple')
plt.fill_between(time, c_low, c_high, color='pink', alpha=0.3)
plt.xlabel('Day')
plt.ylabel('Estimation value')
plt.title('95% confidence interval of c')
plt.legend()
plt.savefig("EM_c.png", dpi=500, bbox_inches = 'tight')
plt.show()
```
